# Define funções

```python
import time
def download_GLM5min_CPTEC(ano, mes, dia, hor, min, dir):

    print('.... Baixando dado ===>', 'DATA=', ano, mes, dia, hor, min)

    # caminho do ftp do cptec
    ftp_cptec_glm_5min = 'http://ftp.cptec.inpe.br/goes/goes16/goes16_web/glm_acumulado_nc/'

    # código do produto
    codigo_produto = 'S11635949'

    # nome do arquivo
    file_glm5min_cptec =  f'{ftp_cptec_glm_5min}{ano}/{mes}/{codigo_produto}_{ano}{mes}{dia}{hor}{min}.nc' # S11635949_202209010000.nc

    # baixando a imagem de satélite
    !wget -c {file_glm5min_cptec} -P {dir}

    # local da imagem que foi baixada
    file_glm_download = f'/content/{codigo_produto}_{ano}{mes}{dia}{hor}{min}.nc'

    return file_glm_download
```

# Processamento

```python
%%time
#============================================================================================================#
#                                      IMPORTANDO BIBLIOTECAS
#============================================================================================================#
from datetime import datetime
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

#============================================================================================================#
#                                       DEFINE O ANO DE DOWNLOAD 
#============================================================================================================#
ano = '2023'

#============================================================================================================#
#                                          CRIANDO AS PASTAS
#============================================================================================================#
# caminho do diretório onde serão salvo os dados
dir = '/home/enrique/projeto_glm/'

# define mês e ano
anoi, mesi, diai = ano, '09', '01'
anof, mesf, diaf = ano, '12', '31'

# data inicial no formato datetime
date_in = datetime(int(anoi), int(mesi), int(diai))
date_ini = date_in.strftime('%Y%m%d')

# data final no formato datetime
date_en = datetime(int(anof), int(mesf), int(diaf))
date_end = date_en.strftime('%Y%m%d')

# loop das pastas por dia
for file in pd.date_range(date_ini, date_end, freq='d'):

    # extrai ano e mes
    ano = file.strftime('%Y')
    mes = file.strftime('%m')
    dia = file.strftime('%d')

    # cria a pasta de saídade com o nome "output"
    os.makedirs(f'{dir}5min/{ano}/{mes}/{dia}', exist_ok=True)
    
#============================================================================================================#
#                                      DOWNLOAD DOS DADOS
#============================================================================================================#
# data inicial no formato datetime
date_in = datetime(int(anoi), int(mesi), int(diai), int(0), int(0))  # 2022-02-02 00:00:00
date_ini = date_in.strftime('%Y%m%d%H%M') # 20220202

# data final no formato datetime
date_en = datetime(int(anof), int(mesf), int(diaf), int(23), int(59))
date_end = date_en.strftime('%Y%m%d%H%M')

# download dos dados
files_imagens = []
for file in pd.date_range(date_ini, date_end, freq='5min'):

    # extrai ano e mes
    data = file.strftime('%Y%m%d%H%M')
    ano = file.strftime('%Y')
    mes = file.strftime('%m')
    dia = file.strftime('%d')
    hor = file.strftime('%H')
    min = file.strftime('%M')

    # baixa dado GLM
    file_glm_download = download_GLM5min_CPTEC(ano, mes, dia, hor, min, f'{dir}5min/{ano}/{mes}/{dia}')
```
